```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(30,20), dpi=100)
```

```python
speedup = [6.225534, 3.245677, 1.634655, 1.091156, 0.8115542]
thread = [2, 4, 8, 12, 16]

plt.xticks(np.arange(2, 17, 2))
plt.title('pi_block_linear', fontsize=18, pad=20)
plt.ylabel('execution time', fontsize=14,rotation=0, labelpad=40)
plt.xlabel('processes', fontsize=14)
for a, b in zip(thread, speedup):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.plot(thread, speedup, 'o-')
plt.show()

```
